```python
#Test de la bibliothèque sklearn
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data-fr/Data_CAC40/dataframe_ACA.PA.csv")
df.head()
```

```python
m = len(df)
Y = df['Close']
X = np.arange(m).reshape(m,1)
```

```python
from sklearn.linear_model import LinearRegression
#Regression linéaire
model = LinearRegression()
model.fit(X,Y)
model.score(X,Y)
```

```python
predic = model.predict(X)
plt.scatter(X,Y, s = 2)
plt.plot(X,predic, c='r')
```

```python
def prediction(t):
    x = np.array([t]).reshape(1,1)
    return model.predict(x)
print(prediction(10*m))
```
